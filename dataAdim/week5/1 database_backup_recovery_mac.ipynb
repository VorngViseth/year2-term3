```python
import os
import subprocess
import datetime

import mysql.connector
from mysql.connector import errorcode
```

```python
def backup_database(host, user, password, db_name, backup_dir):
    # Ensure the backup directory exists
    if not os.path.exists(backup_dir):
        os.makedirs(backup_dir)

    # Create a timestamp for the backup file
    timestamp = datetime.datetime.now().strftime("%Y%m%d%H%M%S")
    backup_file = os.path.join(backup_dir, f"{db_name}_backup1_{timestamp}.sql")

    # Construct the mysqldump command
    dump_command = f"mysqldump -h {host} -u {user} -p{password} {db_name} > {backup_file}"

    try:
        # Execute the backup command
        subprocess.run(dump_command, shell=True, check=True)
        print(f"Backup successful: {backup_file}")
    except subprocess.CalledProcessError as e:
        print(f"Backup failed: {e}")


# backup_dir = 'it create a backup folder in the current directory'
backup_dir = './backups'

# Usage example
backup_database('localhost', 'root', 'root', 'airline_db', backup_dir)
```

```python
import os
import subprocess

def restore_database(host, user, password, db_name, backup_file):
    # Construct the mysql command
    restore_command = f"/Applications/MAMP/Library/bin/mysql -h {host} -u {user} -p{password} {db_name} < {backup_file}"

    try:
        # Execute the restore command
        subprocess.run(restore_command, shell=True, check=True)
        print(f"Restore successful from: {backup_file}")
    except subprocess.CalledProcessError as e:
        print(f"Restore failed: {e}")

# replace this with the actual path to your backup file
backup_dir = './backups/airline_db_backup1_20240528224236.sql'
# Usage example
restore_database('localhost', 'root', 'root', 'airline_db_from_backup', backup_dir)
```
